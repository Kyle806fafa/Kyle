```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Display the column names to identify the correct names
print("Columns in the dataset:", ab.columns)

# Step 1: Adjusting the selected columns based on actual column names in the dataset
# Update the selected columns as per the actual column names present in the dataset
# For example, the column names could be 'List Price', 'Hard_or_Paper', etc.
selected_columns = ['Title', 'List Price', 'Hard_or_Paper']  # Adjust based on actual column names in the dataset

ab_reduced = ab[selected_columns]
ab_reduced_noNaN = ab_reduced.dropna()

# Step 2: Splitting the dataset into an 80/20 train-test split
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Reporting number of observations in train and test sets
print(f"Number of observations in training set: {len(ab_reduced_noNaN_train)}")
print(f"Number of observations in testing set: {len(ab_reduced_noNaN_test)}")

# Step 3: Preparing target variable and feature for training
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']
X = ab_reduced_noNaN_train[['List Price']]

# Step 4: Creating and training the Decision Tree Classifier with max_depth of 2
clf = tree.DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)

# Step 5: Plotting the decision tree
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.show()
```
