# Analysis 1: Hypothesis testing

Research Question: Whether there are statistically significant differences in life satisfaction across different age groups, gender identities, and education levels?

Process: Taking Life satisfaction (scale from 1 to 10) as the dependent variable, and taking age, identity, and education level as independent variables. Then make the hypothesis:  
Null hypothesis (H0): There is no difference in life satisfaction scores among different age groups, identities, and education levels.  
Alternative hypothesis (H1): At least one age group, identity, or education level has a different average life satisfaction score compared to others.  
After that, choose the significance level, which usually takes 0.05.  
Using Python code to compute the corresponding p-value, if the p-value < significance level, reject the null hypothesis.  

Hypothesis: The p-value computed is probably larger than the significance level, suggesting that there is a difference in life satisfaction scores among different age groups, identities, and education levels. 

# Analysis 2: Bootstrapping

Research Question: What is the distribution of life satisfaction mean scores across different independent variables?  

Process: Draw multiple bootstrap samples from the life satisfaction data for each category of age, identity, and education level, then calculate the mean life satisfaction score for each bootstrap sample. After that, using the distribution of bootstrap means to estimate a 95% confidence interval for the mean life satisfaction scores for each group. This method allows us to visualize bootstrap distributions and confidence intervals, and find out how they vary among different independent variables.  

Hypothesis: Probably get a result that reflects a huge difference in life satisfaction mean scores across different independent variables. This means that citizens of Canada with different ages, identities, and education levels have very different scores for life satisfaction.

# Analysis 3: Linear Regression
Research Question: What are the effects of age, identity, and education level on Canadians' life satisfaction scores.

Process: Using the Python code to create a linear regression model, then predict life satisfaction based on age, gender, and education. Also, interpret the coefficients to understand how life satisfaction changes with these variables.

Hypothesis: All three variables have effects on Canadians' life satisfaction scores, but they are different. The coefficient for the age group might be the one that is closest to 1/-1, which means it has the strongest effect. And for the coefficient education level might be the one closest to 0, suggesting it has the smallest effect on life satisfaction scores.
